```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Map, Bar, Page
from pyecharts.globals import ThemeType
```

```python
keyword_list = [keyword for keyword in set(df['keyword'])] 
keyword_groups = df.groupby('keyword')
```

```python
def map_visualmap() -> Page:
    page = Page()
    m = Map({"theme": ThemeType.MACARONS})
    b = Bar({"theme": ThemeType.MACARONS})
    b2 = Bar({"theme": ThemeType.MACARONS})
    
    first = True
    for keyword in keyword_list:
        group = keyword_groups.get_group(keyword)
    
        m.add(keyword, [list(z) for z in zip(group['prov'], group['city_index'])], "china", is_map_symbol_show=False, is_selected=True if first else False)
        
        prov_list = [prov for prov in set(group['prov'])]
        data = [int(group.groupby('prov').get_group(prov)['city_index'].sum()) for prov in prov_list]
        pair = list(zip(prov_list, data))
        pair.sort(key=lambda e: e[1], reverse=True)
        b.add_xaxis([i[0] for i in pair[:10]])
        b.add_yaxis(keyword, [i[1] for i in pair[:10]], is_selected=True if first else False, category_gap="50%")
        b.reversal_axis()
        
        b2.add_xaxis(list(group['city'])[:10])
        b2.add_yaxis(keyword, list(group['city_index'])[:10], is_selected=True if first else False, category_gap="50%")
        b2.reversal_axis()
        
        first = False
        
        
    m.set_global_opts(visualmap_opts=opts.VisualMapOpts(max_=4000),)
    b.set_global_opts(datazoom_opts=opts.DataZoomOpts(range_start=0, range_end=100),)
    b2.set_global_opts(datazoom_opts=opts.DataZoomOpts(range_start=0, range_end=100),)
     
    page.add(m)
    page.add(b)
    page.add(b2)
    
    return page
map_visualmap().render_notebook()
```
